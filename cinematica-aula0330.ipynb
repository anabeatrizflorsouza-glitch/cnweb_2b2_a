#Fórmula da velocidade

```
def calculaVelMedia(deltaS, deltaT):
  velMedia = (deltaS/deltaT)
  return velMedia
```

##Teste Serviço Nuvem: Velocidade Média

```
print("Vm=", calculaVelMedia(30, 15))
```

#Fórmula da aceleração

```
def calculaAceleracaomedia(deltaV, deltaT):
  aceleracaoMedia = (deltaV/deltaT)
  return aceleracaoMedia
```

##Testes de Função : Aceleração Média

```
print("Am=", calculaAceleracaomedia(30,15))
```

#Fórmulas do Movimento Uniforme (MU)

```
def calculaPosicao(posicao_inicial, velocidade, tempo):
  posicao_final = posicao_inicial + (velocidade * tempo)
  return posicao_final
```

##Testes de Função : Posição final

```
print("Movimento uniforme", calculaPosicao(15,2,5))
```

#Função horária do deslocamento

```
def calculaDesloc(inicialS, deltaT, v):
  d = inicialS + (v*deltaT)
  return d
```

##Testes de Função : Horária de Deslocamento

```
print("Deslocamento= ",calculaDesloc(0, 15, 30))
```

#Equação de Torricelli

```
import math

def calTorricelli(vZero, acel, des):
  t=2*acel*des
  v=vZero+t
  res= round (math.sqrt(v) ,2)
  return res
```

##Testes de Função : Torricelli

```
print("Calculo de Torriceli:",calTorricelli(0,3,8))
```
